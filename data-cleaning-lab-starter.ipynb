```python
# Import Modules Here
import pandas as pd
```

## Part 1: Read the Data into Jupyter

```python
# Read the data into a dataframe named pima_df.
# YOUR CODE HERE
pima_df=pd.read_csv('diabetes_uncleaned_data.csv')
```

## Part 2: Missing Data

```python
#Question 1:  Find Columns with missing data
# YOUR CODE HERE
missing_columns = pima_df.columns[pima_df.isna().any()]
print(missing_columns)
```

```python
# Question 2:  Write code to remove rows with missing data and save the new dataframe as pima_missing_fixed_df
# YOUR CODE HERE
pima_missing_fixed_df = pima_df.dropna()

pima_missing_fixed_df.to_csv('pima_missing_fixed.csv')
```

```python
# Question 3:  Print out the number of rows in the dataframe.
# YOUR CODE HERE

print(pima_missing_fixed_df.shape[0])
```

## Part 3: Duplicated Data

```python
# Question 1: find the rows of duplicated data in the 
# pima_missing_fixed_df and print them.  

duplicated_rows = pima_missing_fixed_df[pima_missing_fixed_df.duplicated()]
print(duplicated_rows)
```

```python
# Question 2: Write code to remove rows with missing data 
# and save the new dataframe as pima_dedupped_df

pima_dedupped_df = pima_missing_fixed_df.drop_duplicates()

pima_dedupped_df.to_csv('pima_dedupped.csv')
```

```python
# Question 3:  Print out the number of rows in the dataframe.
print(pima_dedupped_df.shape[0])
```

## Part 4: Mis-typed Columns/Broken Data


```python
# Question 1: find the columns where most of the data seems to be one type of data, and there seems to be a data error. In the markdown, after exploring in code, add a markdown cell discussing which columns have the data error. 
pima_dedupped_df.info()
display(pima_dedupped_df['BloodPressure'].value_counts())
display(pima_dedupped_df['BMI'].value_counts())
display(pima_dedupped_df['Outcome'].value_counts())
```

```python
# Question 2: Remove the rows with data errors
pima_remove_errordf=pima_dedupped_df[(pima_dedupped_df['BloodPressure']!='Error')&(pima_dedupped_df['BMI']!='Error')&(pima_dedupped_df['Outcome']!='ERROR')]

pima_remove_errordf['Insulin'].count()


```

```python
# Question 3: After the data is fixed in your columns change the columns to the correct type and save this as the pima_fixed_columns_df.
pima_fixeddf=pima_remove_errordf.copy()
pima_fixeddf['BloodPressure']=pima_fixeddf['BloodPressure'].astype('int')
pima_fixeddf['BMI']=pima_fixeddf['BMI'].astype('float')

def convert_bool(x):
    print(x)
    if x=="FALSE":
        return False
    elif x=="TRUE":
        return True
    
pima_fixeddf['Outcome']=pima_fixeddf['Outcome'].apply(convert_bool)

```

```python
# Question 4: Run pima_fixed_columns_df.info() to confirm the columns have changed.
pima_fixeddf.info()
display(pima_fixeddf)
```

## Part 5: Outlier Detection and Removal


```python
# Question 1: Print out the Outliers in each column in the pima_fixed_columns_df dataframe, use the IQR method of outlier detection.
def is_iqr_outlier(column):
    q1 = column.quantile(0.25)
    q3 = column.quantile(0.75)
    IQR = q3 - q1
    lower = q1 - (1.5 * IQR)
    higher = q3 + (1.5 * IQR)
    return ~column.between(lower, higher)

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['Pregnancies'])])

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['Glucose'])])

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['BloodPressure'])])

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['SkinThickness'])])

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['Insulin'])])

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['BMI'])])

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['DiabetesPedigreeFunction'])])

display(pima_fixeddf[is_iqr_outlier(pima_fixeddf['Age'])])


```

```python
# Question 2: Use loc to remove outliers in each of the columns that have outliers, save this as pima_outlier_removed_df.

pima_outlier_removed_df = pima_fixeddf.copy()

pima_outlier_removed_df=pima_fixeddf[is_iqr_outlier(pima_fixeddf['Pregnancies'])==False]

print(pima_outlier_removed_df.count())

pima_outlier_removed_df=pima_fixeddf[is_iqr_outlier(pima_fixeddf['BloodPressure'])==False]

print(pima_outlier_removed_df.count())

outlier3df=pima_fixeddf[is_iqr_outlier(pima_fixeddf['Insulin'])==False]

print(outlier3df.count())



pima_outlier_removed_df.to_csv('pima_outlier_removed.csv')
```
